```python
import cv2
import matplotlib.pyplot as plt
from skimage.filters import roberts
import numpy as np

# read image
gambar = cv2.imread('kamera.jpg')

# conversion the image to grayscale
if len(gambar.shape) == 3:  # jika gambar memiliki 3 channel (RGB)
    gambar_gray = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)
else:
    gambar_gray = gambar

# detection edge using operator roberts
tepi_roberts = np.abs(roberts(gambar_gray))

# showing original image and detection edge result
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(tepi_roberts, cmap='gray')
plt.title('Detection Edge Using Operator Roberts')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

#read image
originalImage = cv2.imread('kamera.jpg', cv2.IMREAD_COLOR)

#determine parameter for operator canny
sigma = 1.9 #standard deviation parameter for Gaussian
thresholdLow = 0.05 #lower limit for non-maximum suppression
thresholdHigh = 0.15 #upper limit for non-maximum suppression

#implementation gaussian blur for reduce noise
blurredImage = cv2.GaussianBlur(originalImage, (5,5), sigma)

#detection edge using operator canny
edgeImage = cv2.Canny(blurredImage, thresholdLow * 255, thresholdHigh * 255)

# showing original image and detection edge result
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edgeImage, cmap='gray')
plt.title('Detection Edge Using Canny')
plt.axis('off')

plt.show()
```
